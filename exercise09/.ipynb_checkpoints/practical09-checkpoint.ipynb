# 1 A Medical Probabilistic Graphical Model

```python
def Bernoulli(p): return np.random.binomial(1, p)
# def Categorical(p): return np.random.multinomial(1, p)

def chest_sample():
    a = Bernoulli(0.01)
    s = Bernoulli(0.5)
    t = Bernoulli(0.05) if a else Bernoulli(0.01)
    l = Bernoulli(0.1) if s else Bernoulli(0.01)
    b = Bernoulli(0.6) if s else Bernoulli(0.3)
    e = t or l
    x = Bernoulli(0.98) if e else Bernoulli(0.05)
    if e and b:
        d = Bernoulli(0.9)
    elif e and (not b):
        d = Bernoulli(0.7)
    elif (not e) and b:
        d = Bernoulli(0.8)
    else:
        d = Bernoulli(0.1)
    return a,s,t,l,b,e,x,d
```

```python
chest_sample()
```

```python
N = 100000
data = [chest_sample() for _ in range(N)]
```

```python
# p(d)
count, total = 0,0
for sample in data:
    if sample[7]:
        count += 1
    total += 1
print(f"p(d): {count / total}")
```

```python
# p(d|s=1) and p(d|s=0)
count, total = 0,0
for sample in data:
    if sample[1]:
        total += 1
        if sample[7]:
            count += 1
print(f"p(d|s=1): {count / total}")
# p(d|s=0)
count, total = 0,0
for sample in data:
    if not sample[1]:
        total += 1
        if sample[7]:
            count += 1
print(f"p(d|s=0): {count / total}")
```

```python
# Marginal probability of having lung cancer
count, total = 0,0
for sample in data:
    if True:
        total += 1
        if sample[3]:
            count += 1
print(f"p(l): \t\t{count / total:.4f} \t ({total})")
# Probability of lung cancer giving breathing problems
count, total = 0,0
for sample in data:
    if sample[7]:
        total += 1
        if sample[3]:
            count += 1
print(f"p(l|d): \t{count / total:.4f} \t ({total})")
# Probability of having lung cancer after an additional x-ray
count, total = 0,0
for sample in data:
    if sample[7] and sample[6]:
        total += 1
        if sample[3]:
            count += 1
print(f"p(l|d,x): \t{count / total:.4f} \t ({total})")
# After knowledge of smoking
count, total = 0,0
for sample in data:
    if sample[7] and sample[6] and sample[1]:
        total += 1
        if sample[3]:
            count += 1
print(f"p(l|s,d,x): \t{count / total:.4f} \t ({total})")
# After knowledge of holiday
count, total = 0,0
for sample in data:
    if sample[7] and sample[6] and sample[0]:
        total += 1
        if sample[3]:
            count += 1
print(f"p(l|a,d,x): \t{count / total:.4f} \t ({total})")
# After knowledge of holiday and smoking
count, total = 0,0
for sample in data:
    if sample[7] and sample[6] and sample[0] and sample[1]:
        total += 1
        if sample[3]:
            count += 1
print(f"p(l|a,d,x,s): \t{count / total:.4f} \t ({total})")
```

```python
### Generating random trajectories through the world
# state = 3 # initial state
# circles = [3,5,7]
# n_steps = 30
# seq_act = []
# seq_state = []
# seq_obs = []
# p_sensor = 0.95
# for n in range(n_steps):
#     act = Bernoulli(0.5) # one means move to the right, zero left
#     seq_act.append(act)
#     if act:
#         step = Categorical([0.9, 0.05, 0.05])
#         state += step[0] - step[2]
#     if not act:
#         step = Categorical([0.9, 0.05, 0.05])
#         state += step[2] - step[0]
#     state = state % 10
#     seq_state.append(state)
#     if state in circles:
#         observe = Bernoulli(p_sensor)
#     else:
#         observe = Bernoulli(1 - p_sensor)
#     seq_obs.append(observe)
# print(seq_act)
# print(seq_state)
# print(seq_obs)
# np.savez("hmm-data", aseq=seq_act, rseq=seq_obs)
# np.savez("hmm-data-sol", aseq=seq_act, rseq=seq_obs, lseq=seq_state)
tmp = np.load("hmm-data-sol.npz")
seq_act = tmp["aseq"]
seq_obs = tmp["rseq"]
seq_state = tmp["lseq"]
```

```python
circles = [3,5,7]
p_sensor = 0.95

def update_belief(belief, a, r):
    belief_unnorm = np.zeros(10)
    prx = np.zeros(10)
    for l in range(10):
        # The step
        if a: # Step to the right
            belief_unnorm[l] = 0.9 * belief[(l-1) % 10] + 0.05 * belief[l] + 0.05 * belief[(l+1) % 10] 
        else: # Step to the left
            belief_unnorm[l] = 0.05 * belief[(l-1) % 10] + 0.05 * belief[l] + 0.9 * belief[(l+1) % 10] 
        # Probability of observing a given sensor reading
        if l in circles:
            prx[l] = p_sensor**r * (1 - p_sensor)**(1 - r)
        else:
            prx[l] = (1 - p_sensor)**r * p_sensor**(1 - r)

    noreadout = 1. * belief_unnorm # I.e. before we have seen something
    belief_unnorm *= prx
    belief = belief_unnorm/sum(belief_unnorm)
    return belief, noreadout
```

```python
belief = np.ones(10)/10
for l,a,r in zip(seq_state, seq_act, seq_obs):
    belief, noreadout = update_belief(belief, a, r)
    plt.plot(belief)
    plt.plot(noreadout)
    plt.scatter(l, 0.95)
    sns.despine()
    plt.xlim(0, 9)
    plt.ylim(0, 1)
    plt.title(f"Action: {a}//Sensor: {r}")
    plt.show()
```
