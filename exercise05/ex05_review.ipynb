# 1 Visualize Regularization Contours

```python
tmp = np.load("sheet5-linreg1.npz")
X = tmp['X'] # 2*100
Y = tmp['Y']# 1*100
```

```python
def ridge_reg(beta1, beta2, const):
    return const*(beta1**2 + beta2**2)

def lasso_reg(beta1, beta2, const):
    return const*(abs(beta1) + abs(beta2))

def SSQ(X, Y, beta1, beta2):
    beta = np.vstack((beta1, beta2))
    error = np.sum((Y - beta.T.dot(X))**2, axis=1)
    return error

def SSQ2(X, Y, beta1, beta2):
    beta = np.vstack((beta1, beta2))
    error = []
    for b in beta.T:
        error.append((Y - b.dot(X)).dot((Y - b.dot(X)).T))
    return np.array(error)
```

```python
beta_x = np.arange(-1, 3, 0.1)
beta_y = np.arange(-1, 3, 0.1)
beta_X, beta_Y = np.meshgrid(beta_x, beta_y)
N = beta_X.shape[0]
beta_X = beta_X.reshape(1, -1)
beta_Y = beta_Y.reshape(1, -1)

ridge = ridge_reg(beta_X, beta_Y, 1).reshape(N, -1)
lasso = lasso_reg(beta_X, beta_Y, 1).reshape(N, -1)
SSQ_Error = SSQ(X, Y, beta_X, beta_Y).reshape(N, -1)
SSQ_Error2 = SSQ2(X, Y, beta_X, beta_Y).reshape(N, -1)
beta_X = beta_X.reshape(N, -1)
beta_Y = beta_Y.reshape(N, -1)
plt.contour(beta_X, beta_Y, SSQ_Error, levels=100)
```
