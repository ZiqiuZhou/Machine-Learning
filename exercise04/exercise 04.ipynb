```python
tmp1 = np.load("linear-unc1.npz")
key_X1, key_Y1 = list(tmp1.keys())
features1, targets1= np.matrix(tmp1[key_X1]), list(tmp1[key_Y1])  # 2×10000， 1×10000

tmp2 = np.load("linear-unc2.npz")
key_X2, key_Y2 = list(tmp2.keys())
features2, targets2= np.matrix(tmp2[key_X2]), list(tmp2[key_Y2])  # 2×10000， 1×10000
print(np.shape(features2))

tmp3 = np.load("linear-unc3.npz")
key_X3, key_Y3 = list(tmp3.keys())
features3, targets3= np.matrix(tmp3[key_X3]), list(tmp3[key_Y3])  # 2×10000， 1×10000
```

```python
# for datasets in file "linear-unc1.npz"
bias_x = np.array([[1,1]])  # bias in input data
bias_y = np.array([0])

# for datasets in file "linear-unc1.npz"
subfeatures1 = features1.T
subtargets1 = np.array(targets1).T
subfeatures1 = np.insert(subfeatures1, 0, values=bias_x, axis=0).T  
subtargets1 = np.insert(subtargets1, 0, values=bias_y, axis=0).T 
beta_estimate = np.dot(np.linalg.inv(np.dot(subfeatures1, subfeatures1.T)), np.dot(subfeatures1, subtargets1.T)) 
beta_estimate = np.array(beta_estimate)
print(beta_estimate)

beta_hat = multivariate_normal(beta_estimate.ravel(), np.linalg.inv(np.dot(subfeatures1, subfeatures1.T))) # beta satisfied with normal distribution
X, Y = np.mgrid[0.75:1.25:.1, 1.8:2.2:.1]
pos = np.empty(X.shape + (2,)) 
pos[:, :, 0] = X
pos[:, :, 1] = Y
plt.contourf(X, Y, beta_hat.pdf(pos))
```

```python
tmp = np.load("linear-rel.npz")
key_X, key_Y = list(tmp.keys())
features, targets = np.matrix(tmp[key_X]), list(tmp[key_Y])  # p×N:= 5×506
```

```python
p, N = np.shape(features)
bias_x = np.ones(p)  # bias in input data
bias_y = np.array([0])
features = np.insert(features, 0, values=bias_x, axis=1) # p×N:= 5×507
targets = np.insert(targets, 0, values=bias_y, axis=1) # 1×N:= 1×507
```

```python
beta_origin = np.dot(np.linalg.inv(np.dot(features, features.T)), np.dot(features, targets.T))
residuals_origin = targets - np.dot(beta_origin.T, features)  # 1×507
print(np.linalg.norm(residuals_origin)**2)
p, N = np.shape(features)
```

```python
features_estimate = np.zeros((p, N))
targets_estimate = np.array(targets).ravel()
for i in range(5):
    features_estimate[i] = np.array(features[i]).ravel()
    
features_estimate[0] = np.random.permutation(features_estimate[0]) 
print(features_estimate)
beta = np.dot(np.linalg.inv(np.dot(features_estimate, features_estimate.T)), np.dot(features_estimate, targets_estimate.T))
residuals = targets_estimate - np.dot(beta_origin.T, features_estimate)  # 1×507
print(np.linalg.norm(residuals)**2)
```
