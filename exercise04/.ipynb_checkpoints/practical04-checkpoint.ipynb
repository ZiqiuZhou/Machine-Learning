```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from numpy.linalg import inv
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,10)
```

# Preliminaries

```python
# The two functions we will use throughout the exercise sheet
def fit_beta(X, y, add_bias=True): # Don't do add_bias=False, most of the notebook depends on True!
    # In case numpy steals my dimensions
    if len(X.shape) == 1: 
        X = X[None]
    if len(y.shape) == 1:
        y = y[None]
        
    if add_bias:
        p, N = X.shape
        X = np.vstack([np.ones(N), X])

    beta = inv(X.dot(X.T)).dot(X).dot(y.T) 
    # Have a look at np.linalg.solve for a potentially numerically more stable approach
    # that avoids the manual inversion
    # beta = np.linalg.solve(X.dot(X.T), X.dot(y.T))
    return beta

def predict(beta, X):
    # In case numpy steals my dimensions
    if len(X.shape) == 1: 
        X = X[None]       
    return beta[0] + beta[1:].T.dot(X)

def calc_rss(beta, X, y, ret_res = False):
    # In case numpy steals my dimensions
    if len(X.shape) == 1: 
        X = X[None]
    if len(y.shape) == 1:
        y = y[None]
        
    yhat = predict(beta, X)
    residuals = (y - yhat)
    if ret_res:
        return np.sum(residuals**2), residuals
    else:
        return np.sum(residuals**2)
```

# 1. Plotting of the Residuals

```python
# N = 20
# X = 10*np.random.rand(N) - 5
# Y = 1.5 * X**2 + np.random.randn(N)
# np.savez("linear-res.npz", X=X, Y=Y)
tmp = np.load("linear-res.npz")
X = tmp["X"]
Y = tmp["Y"]
X.shape, Y.shape
# Plot raw data
plt.scatter(X, Y)
sns.despine()
```

```python
# Fit with linear features
beta = fit_beta(X, Y)
Xseq = np.arange(-5,5,0.1)
Yhat = beta[0] + beta[1] * Xseq

Xsq = np.vstack([X, X**2])
betasq = fit_beta(Xsq, Y)
Xseq = np.arange(-5,5,0.1)
Yhatsq = betasq[0] + betasq[1] * Xseq + betasq[2] * Xseq**2
plt.scatter(X, Y)
plt.plot(Xseq, Yhat, label="linear features")
plt.plot(Xseq, Yhatsq, label="squared features")
plt.legend(fontsize=20)
sns.despine()
plt.title("Fit", fontsize=20)
```

```python
# Get residuals for each case
# plt.scatter(X, Y - (beta[0] + beta[1] * X), label="linear features")
plt.scatter(X, Y - (betasq[0] + betasq[1] * X + betasq[2] * X**2), label="squared features")
plt.legend(fontsize=20)
sns.despine()
plt.title("Residuals", fontsize=20)
```

# 2 Uncertainty in the Parameter Estimation

## Data set 1

```python
# N = 10000
# X1 = np.random.randn(N)
# X2 = np.random.randn(N)
# X2 = 0 + 0.1 * np.random.randn(N)
# X2 = -X1 + 0.1 * np.random.randn(N)
# Y = 1 * X1 + 2 * X2 + np.random.randn(N)
# X = np.vstack([X1, X2])
# Y = Y[None]
# np.savez("linear-unc3.npz", X=X, Y=Y)
tmp = np.load("linear-unc1.npz")
X = tmp["X"]
Y = tmp["Y"]
plt.scatter(X[0], X[1], alpha=0.2)

def get_subset(X,Y, N=10):
    indices = np.random.permutation(X.shape[1])[:N]
    return X[:,indices], Y[:,indices]
```

```python
S = 100
for s in range(S):
    N = 10
    Xsub, Ysub = get_subset(X, Y, N=N)
    betahat = fit_beta(Xsub, Ysub, add_bias=False)
    plt.scatter(betahat[0], betahat[1], color="darkblue", alpha=0.4)
    plt.xlabel(r"$\beta_1$")    
    plt.ylabel(r"$\beta_2$")
plt.axhline(2, color="black", linestyle="dashed", alpha=0.8)
plt.axvline(1, color="black", linestyle="dashed", alpha=0.8)
betahat_tot = fit_beta(X, Y, add_bias=False)
plt.title(f"Estimated Parameters = ({betahat_tot[0,0]:.2f}, {betahat_tot[1,0]:.2f})", fontsize=20)
sns.despine()
inv(X.dot(X.T))
```
