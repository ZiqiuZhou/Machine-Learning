```python
import numpy as np
from numpy.linalg import eig
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,10)
```

# 1 Principal Component Analysis
### (a) PCA for visualization

```python
countries = ["Albania", "Austria", "Belgium", "Bulgaria", "Czechoslovakia", "Denmark",
        "E Germany", "Finland", "France", "Greece", "Hungary", "Ireland", "Italy", 
        "Netherlands", "Norway", "Poland", "Portugal", "Romania", "Spain", "Sweden",
        "Switzerland", "UK", "USSR", "W Germany", "Yugoslavia"]
protein_raw = np.loadtxt("protein.txt").T # switch from Nxp to pxN
p, N = protein_raw.shape
print(protein_raw.shape)
assert N == len(countries), "You lost a country somewhere..."
```

```python
X = protein_raw.dot(np.eye(N) - np.ones((N,1)).dot(np.ones((N,1)).T)/N)
# X = protein_raw - protein_raw.mean(1, keepdims=True)
```

```python
# Note how different different version of the same code can perform
tmp_data = np.random.randn(1000, 10000)
one_vec = np.ones((10000, 1))
%timeit tmp_data.dot(np.eye(10000) - one_vec.dot(one_vec.T)/10000)
%timeit tmp_data - tmp_data.mean(1, keepdims=True) # Alternative using broadcasting
```

```python
XXT = X.dot(X.T)# /N
evalue, evector = eig(XXT)
indices = np.argsort(evalue)[::-1]
evalue = evalue[indices]
evector = evector[:,indices]

E2d = evector[:,:2]

protein_2d = E2d.T.dot(X) 

plt.scatter(protein_2d[0,:], protein_2d[1,:])

for i in range(N):
    plt.axes().annotate(countries[i], (protein_2d[0,i] + 0.2, protein_2d[1,i] + 0.2), fontsize=20)
plt.xlabel("PC1", fontsize=20)
plt.ylabel("PC2", fontsize=20)
plt.title("Protein Consumption in Europe", fontsize=20, fontweight='bold')
sns.despine()
plt.show()
```

#### ii) MNIST digits

```python
digits = np.load("digits.npy") / 255 # Scale to [0,1]
plt.imshow(digits[42])
```

```python
dig_vec = digits.reshape(-1,784).T
center = dig_vec.mean(1, keepdims=True)
dig_cent = dig_vec - center
```

```python
XXT = dig_cent.dot(dig_cent.T)
evalue, evector = eig(XXT)
indices = np.argsort(evalue)[::-1]
evalue = evalue[indices]
evector = evector[:,indices]
```

```python
error = evalue[::-1].cumsum()[::-1]
plt.plot(np.arange(1, len(error)+1), error)
plt.ylabel("Error/Distortion", fontsize=20)
plt.xlabel("Nr of Eigenvalues", fontsize=20)
plt.xlim((1,784))
plt.ylim(0)
```

```python
plt.imshow(center.reshape(28,28))
plt.title("The mean three", fontsize=20, fontweight="bold")
```

```python
def mapdownandup(X, Evec, M):
    Esub = Evec[:,:M]
    Xdown = Esub.T.dot(X)
    Xup = Esub.dot(Xdown)
    return Xup
```

```python
Xmapped = mapdownandup(dig_cent, evector, 1)
plt.imshow((Xmapped[:,42] + center[:,0]).reshape(28,28))
plt.title("M=1", fontsize=20, fontweight="bold")
```

```python
Xmapped = mapdownandup(dig_cent, evector, 10)
plt.imshow((Xmapped[:,42] + center[:,0]).reshape(28,28))
plt.title("M=10", fontsize=20, fontweight="bold")
```

```python
Xmapped = mapdownandup(dig_cent, evector, 50)
plt.imshow((Xmapped[:,42] + center[:,0]).reshape(28,28))
plt.title("M=50", fontsize=20, fontweight="bold")
```

```python
Xmapped = mapdownandup(dig_cent, evector, 200)
plt.imshow((Xmapped[:,42] + center[:,0]).reshape(28,28))
plt.title("M=200", fontsize=20, fontweight="bold")
```

```python
Xmapped = mapdownandup(dig_cent, evector, 500)
plt.imshow((Xmapped[:,42] + center[:,0]).reshape(28,28))
plt.title("M=500", fontsize=20, fontweight="bold")
```

# 2 Kernel Density Estimation
### (a) Kernel Implementation

```python
def biweight(x,m,w):
    u = (x - m)/w
    supp = abs(u) < 1
    res = 15/(16*w) * (1 - u**2)**2
    return res * supp
```

```python
xseq = np.arange(-5,5,0.01)
plt.plot(xseq, biweight(xseq, 0, 0.5), label=r"$w=0.5$")
plt.plot(xseq, biweight(xseq, 0, 1), label=r"$w=1$")
plt.plot(xseq, biweight(xseq, 0, 2), label=r"$w=2$")
plt.plot(xseq, biweight(xseq, 0, 3), label=r"$w=3$")
plt.legend(fontsize=20)
plt.title("Biweight kernel", fontsize=20, fontweight="bold")
sns.despine()
```

### (b) Kernel density estimation

```python
data = np.load("samples.npy")
data50 = data[:50]
print(data.shape, data50.shape)
```

```python
[x**2 - y for x in [1,2,3] for y in [3,2,1] if y == 1]
```

```python
def kde(x, obs, w=1):
    return sum([biweight(x, d, w) for d in obs]) / len(obs) # The list comp
```

```python
xseq = np.arange(-10,20,0.001)
#plt.plot(xseq, kde(xseq, data50,0.1), label=r"$w=0.1$")
plt.plot(xseq, kde(xseq, data50,0.5), label=r"$w=0.5$")
plt.plot(xseq, kde(xseq, data50,1), label=r"$w=1$")
plt.plot(xseq, kde(xseq, data50,3), label=r"$w=3$")
plt.plot(xseq, kde(xseq, data50,5), label=r"$w=5$")
plt.legend(fontsize=20)
plt.scatter(data50, np.zeros(len(data50)))
plt.title("Mixture of Normal(-6,1) and truncated (on [0,20]) Cauchy(0, 0.5) ", fontsize=20, fontweight="bold")
sns.despine()
```

### (c) Influence of the samples

```python
S = 50
N = 10
xseq = np.arange(-10,20,0.001)
for i in range(S):
    subdata = data[np.random.randint(0,10000, size=N)]
    # plt.scatter(subdata, np.zeros(len(subdata)))
    plt.plot(xseq, kde(xseq, subdata,1), color="darkblue", alpha=0.2)
plt.title(f"S,N = ({S},{N})", fontsize=20, fontweight="bold")
sns.despine()
```
