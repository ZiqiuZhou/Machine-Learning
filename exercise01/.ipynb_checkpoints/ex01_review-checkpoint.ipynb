```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interactive
plt.rcParams["figure.figsize"] = (20,10)
```

# 1 Principal Component Analysis

```python
data = np.loadtxt("data_ex01/protein.txt")
countries = ["Albania", "Austria", "Belgium", "Bulgaria", "Czechoslovakia", "Denmark",
        "E Germany", "Finland", "France", "Greece", "Hungary", "Ireland", "Italy", 
        "Netherlands", "Norway", "Poland", "Portugal", "Romania", "Spain", "Sweden",
        "Switzerland", "UK", "USSR", "W Germany", "Yugoslavia"]

n = data.shape[0]
p = data.shape[1]
data # N:25 p: 9
```

```python
center_Hat = np.identity(n) - (1 / n) * np.ones((n, n))
data_centered = data.T.dot(center_Hat)  # 9*25
```

# (a) PCA for visualization

```python
eigenvalue, eigenvector = np.linalg.eig(data_centered.dot(data_centered.T))
indics = eigenvalue.argsort()[::-1]
eigenvector_sort = eigenvector[:, indics]
pc2 = eigenvector_sort[:, : 2]  # first 2 principle components
```

```python
score = pc2.T.dot(data_centered)
plt.scatter(score[0], score[1])

for i in range(n):
    plt.annotate(countries[i], xy=(score[0, i] + 0.2, score[1, i] + 0.2), fontsize=15)
plt.xlabel("pc1", fontsize=15)
plt.ylabel("pc2", fontsize=15)
plt.title("Protein Consumption in Europe", fontsize=20, fontweight='bold')
sns.despine()
```

# (b) PCA for lossy compression

```python
plt.plot(np.arange(p), eigenvalue)
plt.xlabel("dimensions", fontsize=15)
plt.ylabel("eigenvalues", fontsize=15)
sns.despine()
```

```python
Error = []
for m in range(p):
    error = np.sum(eigenvalue[m+1::])
    Error.append(error)
plt.plot(np.arange(1,p+1), Error)
```

```python
digits = np.load("data_ex01/digits.npy") / 255  # Scale to [0,1] with 6131*28*28
plt.imshow(digits[666])
digits.shape
```

```python
from scipy import ndimage

data_raw = np.reshape(digits, (digits.shape[0], -1)).T # 784*6131
N = data_raw.shape[1]
#center_Hat = np.identity(N) - (1 / N) * np.ones((N, N))
#data_centered = data_raw.dot(center_Hat)
center = data_raw.mean(1, keepdims=True)  # 784
print(np.shape(center))
dig_cent = data_raw - center

def g(M):
    eigenvalue, eigenvector = np.linalg.eig(dig_cent.dot(dig_cent.T))
    indics = np.argsort(eigenvalue)[::-1]
    eigenvector_sort = eigenvector[:, indics]
    w = eigenvector_sort[:, :M]
    approximation_data = w.dot(w.T).dot(dig_cent)
    approximation_data = approximation_data + center
    approximation_image = np.reshape(approximation_data.T, (-1, 28, 28))
    plt.imshow(approximation_image[666])
    
interactive(g, M=(1, 784, 50))
```

# 2 Kernel Density Estimation

```python
def kernel(x, m, w):
    supp = abs((x - m) / w) <= 1
    k = (15 / (16 * w)) * (1 - ((x - m) / w)**2)**2
    return k * supp
```

```python
x = np.arange(-1, 1, 0.01)
biweight_kernel = kernel(x, 0, 1)
plt.plot(x, kernel(x, 0, 1), label=r"$w=1$")
plt.plot(x, kernel(x, 0, 0.3), label=r"$w=0.3$")
plt.plot(x, kernel(x, 0, 0.5), label=r"$w=0.5$")
plt.plot(x, kernel(x, 0, 0.1), label=r"$w=0.1$")
plt.legend(fontsize=15)
```

# (b) Kernel density estimation

```python
data = np.load("data_ex01/samples.npy")  # 10000
data_sampled = data[:50]
size = data_sampled.shape[0]  # 50
```

```python
data_range = np.arange(-10, 20, 0.01)

def g(N):
    kde = []
    for x in data_sampled:
        k = kernel(data_range, x, N)
        kde.append(k)
    kde = np.reshape(kde, (size, -1))
    kde = np.sum(kde, axis=0) / size
    plt.plot(data_range, kde, alpha=0.5)
    plt.xlabel("range", fontsize=15)
    plt.ylabel("kde", fontsize=15)
    plt.title("Kernel Density Estimation", fontsize=20, fontweight='bold')
    sns.despine()
interactive(g, N=[0.1, 0.5, 1, 3, 5])
```

```python
def kernel_density_estimation(data_sampled, data_range, w):
    kde = []
    size = data_sampled.shape[0]
    for x in data_sampled:
        k = kernel(data_range, x, w)
        kde.append(k)
    kde = np.reshape(kde, (size, -1))
    kde = np.sum(kde, axis=0) / size
    plt.plot(data_range, kde, alpha=0.3)  
```

```python
bandwidth = 1
for i in range(100, 10000, 500):
    sampled_data = np.random.permutation(data)[:i]
    kernel_density_estimation(sampled_data, data_range, bandwidth)
    
plt.xlabel("range", fontsize=15)
plt.ylabel("kde", fontsize=15)
plt.title("Kernel Density Estimation", fontsize=20, fontweight='bold')
sns.despine()
```
